### removing html tags

```python
import re
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r"",text)
```

```python
df['review'] = df['review'].apply(remove_html_tags)
```

```python
import string
exclude = string.punctuation
```

```python
def remove_punc(text):
    for char in exclude:
        text = text.replace(char,"")
    return text
```

```python
df['review'] = df['review'].apply(remove_punc)
```

```python
df['review'][1]
```

```python
chat_abbreviations = {
    'afaik': 'as far as I know',
    'brb': 'be right back',
    'btw': 'by the way',
    'omg': 'oh my god',
    'lol': 'laughing out loud',
    'idk': 'I dont know',
    'tbh': 'to be honest',
    'smh': 'shaking my head',
    'fyi': 'for your information',
    'lmao': 'laughing my ass off',
    'imo': 'in my opinion',
    'imho': 'in my humble opinion',
    'rofl': 'rolling on the floor laughing',
    'nvm': 'never mind',
    'ttyl': 'talk to you later',
    'wyd': 'what you doing',
    'wtf': 'what the f*ck',
    'wth': 'what the hell',
    'ftw': 'for the win',
    'glhf': 'good luck have fun',
    'yolo': 'you only live once',
    'gtg': 'got to go',
    'np': 'no problem',
    'lmk': 'let me know',
    'bbl': 'be back later',
}
```

```python
def expand_chat_abbreviations(text):
    words = text.split()
    expanded_words = [chat_abbreviations.get(word.lower(), word) for word in words]
    return ' '.join(expanded_words)
```

```python
df['review'] = df['review'].apply(expand_chat_abbreviations)
```

```python
import emoji
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
def preprocess_text(text):
    if pd.isna(text):  # Check for NaN values
        return ""  # Return empty string for missing values
    text = text.lower()  # Lowercasing
    text = BeautifulSoup(text, "html.parser").get_text()  # Remove HTML tags
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  # Removing URLs
    text = emoji.replace_emoji(text, replace='')  # Removing emojis
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Removing special characters and punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # Removing extra spaces
    tokens = word_tokenize(text)  # Tokenization using NLTK
    stop_words = set(stopwords.words('english'))  # Load stopwords
    tokens = [word for word in tokens if word not in stop_words]  # Stopword removal
    lemmatizer = WordNetLemmatizer()  # Initialize lemmatizer
    tokens = [lemmatizer.lemmatize(word) for word in tokens]  # Lemmatization
    return ' '.join(tokens)
```

```python
df_sample['sentiment'] = df_sample['sentiment'].map({
    "positive": 1,
    "negative": 0
})
```

```python
df_sample['cleaned_review'] = df_sample['review'].apply(preprocess_text)
```

```python
df_sample.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
x = df_sample['review']
y = df_sample['sentiment']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.35,random_state = 42)
```

```python
x_train.head()
```

```python
vec = TfidfVectorizer(max_features = 5000)
x_train_tfidf = vec.fit_transform(x_train)
x_test_tfidf = vec.fit_transform(x_test)
```

### model definitions

```python
logreg_model = LogisticRegression()
svm_model = SVC()
nb_model = MultinomialNB()
```

#### Logistic regression

```python
logreg_model.fit(x_train_tfidf,y_train)
```

```python
y_pred_logreg = logreg_model.predict(x_test_tfidf)
print(y_pred_logreg)
```

```python
print("Logistic Regression Accuracy: ", accuracy_score(y_test, y_pred_logreg))
print("Classification Report: \n", classification_report(y_test, y_pred_logreg))
```

#### Naive Bayes

```python
nb_model.fit(x_train_tfidf, y_train)
```

```python
y_pred_nb = nb_model.predict(x_test_tfidf)
print(y_pred_nb)
```

```python
print("Naive Bayes Accuracy: ", accuracy_score(y_test, y_pred_nb))
print("Classification Report: \n", classification_report(y_test, y_pred_nb))
```

#### SVM

```python
svm_model.fit(x_train_tfidf,y_train)
```

```python
y_pred_svm = svm_model.predict(x_test_tfidf)
print(y_pred_svm)
```

```python
print("SVM Accuracy: ", accuracy_score(y_test, y_pred_svm))
print("Classification Report: \n", classification_report(y_test, y_pred_svm))
```
